```python
def calculate_bmi(weight, height):
    """
    計算BMI指數的函式

    Args:
    weight (float): 以公斤為單位的體重
    height (float): 以公尺為單位的身高

    Returns:
    float: BMI指數
    """
    bmi = weight / (height ** 2)
    return round(bmi, 2)  # 將BMI指數保留兩位小數

def interpret_bmi(bmi):
    """
    解釋BMI指數的函式

    Args:
    bmi (float): BMI指數

    Returns:
    str: BMI指數的解釋
    """
    if bmi < 18.5:
        return "過輕"
    elif bmi >= 18.5 and bmi < 24:
        return "正常"
    elif bmi >= 24 and bmi < 27:
        return "過重"
    elif bmi >= 27 and bmi < 30:
        return "輕度肥胖"
    elif bmi >= 30 and bmi < 35:
        return "中度肥胖"
    else:
        return "重度肥胖"

def get_valid_weight(prompt):
    """
    提示用戶輸入並確保體重是有效的數值

    Args:
    prompt (str): 提示字串

    Returns:
    float: 有效的數值
    """
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                raise ValueError("請輸入一個大於零的數字。")
            return value
        except ValueError:
            print("請輸入有效的數字。")

def get_valid_height(prompt):
    """
    提示用戶輸入並確保身高是有效的數值（以公分或是公尺為單位）

    Args:
    prompt (str): 提示字串

    Returns:
    float: 有效的數值（以公尺為單位）
    """
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                raise ValueError("請輸入一個大於零的數字。")
            if value > 3:  # 如果輸入值大於3，假設使用者輸入的是公分而非公尺
                value /= 100  # 將公分轉換成公尺
            return value
        except ValueError:
            print("請輸入有效的數字。")

# 輸入姓名
name = input("請輸入您的姓名：")

# 輸入體重和身高
weight = get_valid_weight("請輸入您的體重（公斤）：")
height = get_valid_height("請輸入您的身高（公分或公尺）：")

# 計算BMI指數
bmi = calculate_bmi(weight, height)

# 解釋BMI指數
interpretation = interpret_bmi(bmi)

print("{} 先生/小姐，您的BMI指數為：{}".format(name, bmi))
print("您的體重狀況為：", interpretation)
```

```python
def calculate_bmi(weight, height):
    return round(weight / (height ** 2), 2)

def interpret_bmi(bmi):
    if bmi < 18.5: return "過輕"
    elif bmi < 24: return "正常"
    elif bmi < 27: return "過重"
    elif bmi < 30: return "輕度肥胖"
    elif bmi < 35: return "中度肥胖"
    else: return "重度肥胖"

name = input("請輸入您的姓名：")
while True:
    try:
        weight = float(input("請輸入您的體重（公斤）："))
        if weight <= 0:
            raise ValueError("請輸入大於零的數字。")
        break
    except ValueError:
        print("請輸入有效的數字。")

while True:
    try:
        height_input = input("請輸入您的身高（公分或公尺）：")
        height = float(height_input)
        if height <= 0:
            raise ValueError("請輸入大於零的數字。")
        if height > 3:
            height /= 100
        break
    except ValueError:
        print("請輸入有效的數字。")

bmi = calculate_bmi(weight, height)
interpretation = interpret_bmi(bmi)

print("{} 先生/小姐，您的BMI指數為：{}".format(name, bmi))
print("您的體重狀況為：", interpretation)
```
